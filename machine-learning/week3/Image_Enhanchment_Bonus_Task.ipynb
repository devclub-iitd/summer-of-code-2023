Image Enhancement Function
Next, we will define a Python function named enhance_image_from_file that takes the file path of an image as input and performs the enhancement. The function will apply histogram equalization to each color channel (blue, green, and red) and then use CLAHE to further enhance the image.


````python
!pip install opencv-python
````

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt
````

````python
def enhance_image_from_file(image_path):
    # Load the image from the local file path
    img = cv2.imread(image_path, cv2.IMREAD_COLOR)

    # Split the image into its color channels
    b, g, r = cv2.split(img)

    # Perform Histogram Equalization on each channel
    b_equalized = cv2.equalizeHist(b)
    g_equalized = cv2.equalizeHist(g)
    r_equalized = cv2.equalizeHist(r)

    # Perform CLAHE (Contrast Limited Adaptive Histogram Equalization) on each channel
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    b_clahe = clahe.apply(b_equalized)
    g_clahe = clahe.apply(g_equalized)
    r_clahe = clahe.apply(r_equalized)

    # Merge the enhanced color channels back into a BGR image
    enhanced_img = cv2.merge([b_clahe, g_clahe, r_clahe])

    # Display the images
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.title('Original Image')
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

    plt.subplot(1, 2, 2)
    plt.title('Enhanced Image')
    plt.imshow(cv2.cvtColor(enhanced_img, cv2.COLOR_BGR2RGB))

    plt.show()
````

````python
# Example usage with a local image in Google Colab:
image_path = 'cloud.jpeg'
enhance_image_from_file(image_path)
````

````python
# Example usage with a local image in Google Colab:
image_path = 'river.jpeg'
enhance_image_from_file(image_path)
````
