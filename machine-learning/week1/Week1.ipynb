```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("C:\\Users\\91740\\Downloads\\dmvdvn.csv")
df
```

```python
a=pd.get_dummies(df.brand)
a
```

```python
b=pd.get_dummies(df.categories)
b
```

```python
df.rename(columns={'prices.merchant':'platform'},inplace=True)
df
```

```python
c=pd.get_dummies(df.platform)
c
```

```python
c.drop('1win',axis="columns",inplace=True)
```

```python
X=pd.concat([a,b,c],axis="columns")
X
```

```python
Y=df.price
Y
```

```python
from sklearn import linear_model
reg=linear_model.LinearRegression()
```

```python
reg.fit(X,Y)
```

```python
reg.score(X,Y)
```
